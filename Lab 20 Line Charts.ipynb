### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
age_groups = {
    'Under 18 years old': 0,
    '18-24 years old': 1,
    '25-34 years old': 2,
    '35-44 years old': 3,
    '45-54 years old': 4,
    '55-64 years old': 5,
    '65 years or older': 6,
    'Prefer not to say': None
}

df['AgeGroup'] = df['Age'].map(age_groups)

df = df.dropna(subset=['AgeGroup'])

age_compensation = df.groupby('AgeGroup')['ConvertedCompYearly'].median()

age_labels = ['Under 18', '18-24', '25-34', '35-44', '45-54', '55-64', '65+']

plt.figure(figsize=(10, 6))
plt.plot(age_labels, age_compensation, marker='o', linestyle='-', color='b')
plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
age_groups = {
    'Under 18 years old': 0,
    '18-24 years old': 1,
    '25-34 years old': 2,
    '35-44 years old': 3,
    '45-54 years old': 4,
    '55-64 years old': 5,
    '65 years or older': 6,
    'Prefer not to say': None
}

df['AgeGroup'] = df['Age'].map(age_groups)

df = df.dropna(subset=['AgeGroup'])

df_filtered = df[df['AgeGroup'].isin([2, 3, 4])]

age_compensation_filtered = df_filtered.groupby('AgeGroup')['ConvertedCompYearly'].median()

age_labels_filtered = ['25-34', '35-44', '45-54']

plt.figure(figsize=(10, 6))
plt.plot(age_labels_filtered, age_compensation_filtered, marker='o', linestyle='-', color='g')
plt.title('Median Yearly Compensation for Ages 25-45')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Convert 'YearsCode' to numeric
df['WorkExp'] = df['WorkExp'].replace({
    'Less than 1 year': 0,
    'More than 50 years': 51
}).astype(float)  # Convert to float (handles NaN)

# Drop NaN values in 'YearsCode' and 'JobSatPoints_6'
df = df.dropna(subset=['WorkExp', 'JobSatPoints_6'])

# Group by experience (YearsCode) and calculate median job satisfaction
exp_satisfaction = df.groupby('WorkExp')['JobSatPoints_6'].median()

# Plot the line chart
plt.figure(figsize=(12, 6))
plt.plot(exp_satisfaction.index, exp_satisfaction.values, marker='o', linestyle='-', color='g')
plt.title('Median Job Satisfaction by Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Median Job Satisfaction (JobSatPoints_6)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
df['WorkExp'] = pd.to_numeric(df['WorkExp'], errors='coerce')

df = df.dropna(subset=['WorkExp', 'ConvertedCompYearly'])

exp_compensation = df.groupby('WorkExp')['ConvertedCompYearly'].median()

plt.figure(figsize=(12, 6))
plt.plot(exp_compensation.index, exp_compensation.values, marker='o', linestyle='-', color='b')
plt.title('Median Compensation by Years of Work Experience')
plt.xlabel('Years of Work Experience')
plt.ylabel('Median Yearly Compensation (USD)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Convert 'WorkExp' and 'JobSatPoints_6' to numeric (if not already)
df['WorkExp'] = pd.to_numeric(df['WorkExp'], errors='coerce')
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Drop NaN values in 'WorkExp' and 'JobSatPoints_6'
df = df.dropna(subset=['WorkExp', 'JobSatPoints_6'])

# Group by experience (WorkExp) and calculate median job satisfaction
exp_satisfaction = df.groupby('WorkExp')['JobSatPoints_6'].median()

# Plot the line chart
plt.figure(figsize=(12, 6))
plt.plot(exp_satisfaction.index, exp_satisfaction.values, marker='o', linestyle='-', color='g')
plt.title('Median Job Satisfaction by Years of Work Experience')
plt.xlabel('Years of Work Experience')
plt.ylabel('Median Job Satisfaction (Scale: 1-10)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
